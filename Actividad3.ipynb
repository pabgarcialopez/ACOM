```python
# Funciones auxiliares
def adecuar_matrices(A, B, n):
  """
  Devuelve las nuevas matrices habiendoles a√±adido una columna y una fila de 0s.
  """

  # A√±adir una columna y una fila de ceros en A y en B
  A = np.insert(np.insert(A, n, 0, 0), n, 0, 1)
  B = np.insert(np.insert(B, n, 0, 0), n, 0, 1)

  return A, B

def obtener_submatrices(M):
  """
  Devuelve los bloques cuadrados que componen la matriz M.
  Necesario que M tenga tama√±o par para funcionamiento correcto.
  """

  half = len(M) // 2  # Importante la division entera por tema de √≠ndices

  return M[:half, :half], M[:half, half:], M[half:, :half], M[half:, half:]
```

```python
# Funciones principales

def mul_mat_escuela(A, B, n):
  """
  Devuelve la matriz producto A * B.
  A y B deben ser ambas matrices n x n
  """
  return np.array([[sum(A[i][k] * B[k][j] for k in range(n)) for j in range(n)] for i in range(n)])

def mul_mat_strassen(A, B, n):

  # Caso base
  if n <= 5:
    return mul_mat_escuela(A, B, n)

  # Caso recursivo

  # Adecuar si es necesario las matrices.
  adecuar = (n % 2 != 0)

  if adecuar:
    A, B = adecuar_matrices(A, B, n)
    n += 1

  # Descomponer por bloques A y B
  A11, A12, A21, A22 = obtener_submatrices(A)
  B11, B12, B21, B22 = obtener_submatrices(B)

  n_2 = n // 2

  # Calcular los valores de Strassen
  M1 = mul_mat_strassen(A11 + A22, B11 + B22, n_2)
  M2 = mul_mat_strassen(A21 + A22, B11, n_2)
  M3 = mul_mat_strassen(A11, B12 - B22, n_2)
  M4 = mul_mat_strassen(A22, B21 - B11, n_2)
  M5 = mul_mat_strassen(A11 + A12, B22, n_2)
  M6 = mul_mat_strassen(A21 - A11, B11 + B12, n_2)
  M7 = mul_mat_strassen(A12 - A22, B21 + B22, n_2)

  # Calcular los bloques resultantes
  C11 = M1 + M4 - M5 + M7
  C12 = M3 + M5
  C21 = M2 + M4
  C22 = M1- M2 + M3 + M6

  C = conc((conc((C11, C12), axis=1), conc((C21, C22), axis=1)), axis=0)

  return C[:-1, :-1] if adecuar else C
```

```python
# M√©todo de la escuela
def mul_mat_escuela(A, B, n):
    """
    Devuelve la matriz producto A * B.
    A y B deben ser ambas matrices n x n
    """

    return [[sum(A[i][k] * B[k][j] for k in range(n)) for j in range(n)] for i in range(n)]

# Funci√≥n auxiliar para sumar matrices
def suma_matrices(A, B):
    n = len(A)
    return [[A[i][j] + B[i][j] for j in range(n)] for i in range(n)]

# Funci√≥n auxiliar para restar matrices
def resta_matrices(A, B):
    n = len(A)
    return [[A[i][j] - B[i][j] for j in range(n)] for i in range(n)]

# Funci√≥n para dividir una matriz en 4 submatrices
def dividir_matriz(M):
    n = len(M)
    mid = n // 2
    A11 = [row[:mid] for row in M[:mid]]
    A12 = [row[mid:] for row in M[:mid]]
    A21 = [row[:mid] for row in M[mid:]]
    A22 = [row[mid:] for row in M[mid:]]
    return A11, A12, A21, A22

# Funci√≥n auxiliar para unir 4 submatrices en una sola
def unir_matrices(A11, A12, A21, A22):
    n = len(A11) * 2
    C = [[0] * n for _ in range(n)]
    mid = n // 2
    for i in range(mid):
        for j in range(mid):
            C[i][j] = A11[i][j]
            C[i][j + mid] = A12[i][j]
            C[i + mid][j] = A21[i][j]
            C[i + mid][j + mid] = A22[i][j]
    return C

# M√©todo de Strassen
def mul_mat_strassen(A, B, n):
    if n <= 5:
        return mul_mat_escuela(A, B, n)

    # Si n es impar, extendemos las matrices con ceros
    if n % 2 != 0:
        A = [row + [0] for row in A] + [[0] * (n + 1)]
        B = [row + [0] for row in B] + [[0] * (n + 1)]

    # Dividir las matrices A y B en 4 submatrices
    A11, A12, A21, A22 = dividir_matriz(A)
    B11, B12, B21, B22 = dividir_matriz(B)

    # Calcular los productos intermedios utilizando las 7 multiplicaciones de Strassen
    M1 = mul_mat_strassen(suma_matrices(A11, A22), suma_matrices(B11, B22), len(A11))
    M2 = mul_mat_strassen(suma_matrices(A21, A22), B11, len(A11))
    M3 = mul_mat_strassen(A11, resta_matrices(B12, B22), len(A11))
    M4 = mul_mat_strassen(A22, resta_matrices(B21, B11), len(A11))
    M5 = mul_mat_strassen(suma_matrices(A11, A12), B22, len(A11))
    M6 = mul_mat_strassen(resta_matrices(A21, A11), suma_matrices(B11, B12), len(A11))
    M7 = mul_mat_strassen(resta_matrices(A12, A22), suma_matrices(B21, B22), len(A11))

    # Combinar los resultados intermedios para formar las submatrices del resultado final
    C11 = suma_matrices(resta_matrices(suma_matrices(M1, M4), M5), M7)
    C12 = suma_matrices(M3, M5)
    C21 = suma_matrices(M2, M4)
    C22 = suma_matrices(resta_matrices(suma_matrices(M1, M3), M2), M6)

    # Unir las submatrices en una matriz grande
    C = unir_matrices(C11, C12, C21, C22)

    # Si extendimos las matrices, eliminamos la fila y columna extra
    if len(C) > n:
        C = [row[:n] for row in C[:n]]

    return C
```
